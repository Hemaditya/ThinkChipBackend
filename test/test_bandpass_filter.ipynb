```python
import filters as f
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a sine wave with multiple frequencies
time = np.arange(0,1,1/250.0)

sine_wave_50hz = np.sin(2*np.pi*50*(time))
sine_wave_25hz = np.sin(2*np.pi*25*(time))
sine_wave_15hz = np.sin(2*np.pi*15*(time))
sine_wave_8hz = np.sin(2*np.pi*8*(time))
sine_wave_5hz = np.sin(2*np.pi*5*(time))

combined_sine_wave = sine_wave_50hz+sine_wave_25hz+sine_wave_15hz+sine_wave_5hz+sine_wave_8hz
plt.plot(time,combined_sine_wave)
```

```python
# Now filter from 23-26hz
fData = (f.apply_bandpass_filter(combined_sine_wave.reshape(1,1,250),23,26))
```

```python
fig,ax = plt.subplots(2,2)
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=0.5)
ax[0,0].plot(combined_sine_wave)
ax[0,0].set_title("all sine_waves Combined Sine Wave")
ax[1,0].plot(fData.reshape(-1),)
ax[1,0].set_title(label="23-26hz filtered Sine Wave")
ax[0,1].plot(sine_wave_50hz.reshape(-1),)
ax[0,1].set_title(label="50hz Sine Wave")
ax[1,1].plot(sine_wave_25hz,)
ax[1,1].set_title(label="25hz Sine Wave")
```
